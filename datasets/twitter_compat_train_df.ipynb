```python
df_train = pd.read_csv('./imbalanced_train_id')
print(df_train.shape)
df_train.head()
```

```python
cat_no_dict = {}

for i in range(df_train.shape[0]):
    cat_no = df_train.iloc[i,4]
    category = df_train.iloc[i,3]
    emoji = df_train.iloc[i,2]
    
    if cat_no not in cat_no_dict: 
        cat_no_dict[cat_no] = []
                        
    if emoji not in cat_no_dict[cat_no]:
        cat_no_dict[cat_no].append(emoji) 
```

```python
df_emoji = pd.DataFrame(cat_no_dict.items(), columns=['Category_No', 'Emojis'])

df_emoji["Category"] = np.nan
df_emoji
```

```python
df_emoji.loc[df_emoji.index[0], 'Category'] = "People & Body" #cat 1--done
df_emoji.loc[df_emoji.index[1], 'Category'] = "Smileys & Emotion" #cat 0--done
df_emoji.loc[df_emoji.index[2], 'Category'] = "Unknown" #cat 7--done
df_emoji.loc[df_emoji.index[3], 'Category'] = "Travel & Places" #cat 9--done
df_emoji.loc[df_emoji.index[4], 'Category'] = "Animals & Nature" #cat 2--done
df_emoji.loc[df_emoji.index[5], 'Category'] = "Symbols" #cat 3---done
df_emoji.loc[df_emoji.index[6], 'Category'] = "Flags" #cat 6---done
df_emoji.loc[df_emoji.index[7], 'Category'] = "Activities" #cat 5---done
df_emoji.loc[df_emoji.index[8], 'Category'] = "Objects" #cat 4---done
df_emoji.loc[df_emoji.index[9], 'Category'] = "Food & Drink" #cat 8
```

```python
df_emoji.to_csv('./emoji_train.csv', index=False)
```
